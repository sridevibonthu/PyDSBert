```python
import pandas as pd
df = pd.DataFrame(dataset['train'])
```

```python
len(df)
```

```python
df.head(2)
```

```python
df['textlength'] = df['text'].apply(len)
```

```python
df['wordsintext'] = df['text'].str.split().str.len()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
sns.set_theme(style='white', 
              palette='dark', 
              rc={'figure.dpi': 300, 'figure.figsize': (5, 3.75)}, 
              font_scale=0.65)
```

```python
sns.histplot(x='textlength', data=df, kde=True)
plt.xlabel("Text length in characters.")
plt.ylabel("Frequency")
#plt.show()
plt.savefig("chardistribution.png")
```

```python
sns.histplot(x='wordsintext', data=df, kde=True)
plt.xlabel("Text length in words.")
plt.ylabel("Frequency")
#plt.show()
plt.savefig("worddistribution.png")
```
